```python
import pandas as pd
import numpy as np

df_bikes = pd.read_csv('../input/data-wrangling-bike-rentals/bike_rentals_cleaned.csv')

df_bikes.head()
```

```python
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=2)
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error',cv=5)
rmse = np.sqrt(-scores)
print('RMSE:',np.round(rmse,2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

```python
from sklearn.metrics import mean_squared_error

dec_reg = DecisionTreeRegressor()

dec_reg.fit(X_train,y_train)

y_preds = dec_reg.predict(X_train)

mse = mean_squared_error(y_preds,y_train)

mse = np.sqrt(mse)

mse

```

A RMSE of means that the model has perfectly fit every data point! This perfect score combined with a cross-validation error 1233.36 is proof that the decision tree is overfitting the data with high variance. The training set fit perfectly, but the test set missed badly.
Hyperparameters may rectify the situation.
